```python
from transformers import pipeline

qa = pipeline('question-answering', 
              model="Josue/BETO-espanhol-Squad2", 
              tokenizer="Josue/BETO-espanhol-Squad2")
```

```python
import base64
import requests

dataset = "https://raw.githubusercontent.com/josuemzx/samantha-IA/main/datasetSIA.txt"
dataset = requests.get(dataset)
dataset = dataset.text
print(dataset)
```

Ejemplo de preguntas:

* Cómo puedo ser reconocido como refugiado? 
* Qué trámites debo realizar para ser reconocido como refugiado?
*	En qué consiste el derecho de asilo o derecho de acogida? 

```python
pregunta= "En qué consiste el derecho de asilo o derecho de acogida"
```

```python
resp = qa(context=dataset, question=pregunta)

print("Pregunta:", pregunta)
print("Respuesta:", resp['answer'])
print("Score:", resp['score'])
```
